### 4.3 Eigenvalues & Eigenvectors

For the matrix:
```
A = [[4, 1],
     [2, 3]]
```

1. Compute eigenvalues and eigenvectors using `np.linalg.eig`.
2. **Verify** the eigenvector equation: `A @ v = λ * v` for each eigenpair.
3. Plot both eigenvectors as arrows, scaled by their eigenvalues, on a 2D plot.
4. **Explain** in a markdown cell (below): what does it geometrically mean for a matrix to stretch a vector?

````python
A = np.array([[4, 1], [2, 3]], dtype=float)

eigenvalues, eigenvectors = np.linalg.eig(A)  # YOUR CODE HERE

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
# YOUR CODE HERE
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]
    Av = A @ v
    lam_v = lam * v
    is_equal = np.allclose(Av, lam_v)
    print(f"Eigenpair {i+1} verification (Av = λv): {is_equal}")

# Plot
# YOUR CODE HERE
fig, ax = plt.subplots()
v1_scaled = eigenvectors[:, 0] * eigenvalues[0]
v2_scaled = eigenvectors[:, 1] * eigenvalues[1]
ax.quiver(0, 0,v1_scaled[0], v1_scaled[1],color='red', angles='xy', scale_units='xy', scale=1, label=f'λ1 = {eigenvalues[0]:.1f}')
ax.quiver(0, 0, v2_scaled[0], v2_scaled[1], color='blue', angles='xy', scale_units='xy', scale=1, label=f'λ2 = {eigenvalues[1]:.1f}')
ax.set_xlim(-5, 6)
ax.set_ylim(-5, 6)
ax.axhline(0, color='k', lw=0.8)
ax.axvline(0, color='k', lw=0.8)
ax.grid(True, linestyle='--', alpha=0.6)
ax.set_aspect('equal') # Ensures 1 unit on X is the same visual length as 1 unit on Y
ax.set_title('Eigenvectors Scaled by Eigenvalues')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend()
plt.show()

print("4.3 passed ✓")
````

**Your Geometric Explanation (4.3):**

When you multiply a vector by a matrix, it usually changes the vector's direction and length. But some special vectors, called eigenvectors, don't get rotated at all. They stay on their exact same line.

The eigenvalue is just the number that tells us what happens to that specific eigenvector's length:

* If the eigenvalue is 2, the vector gets stretched to twice its size.

* If it's 0.5, it gets squished to half its size.

* If it's negative, it just flips around to point the exact opposite way.


So basically, when we say a matrix "stretches" a vector, it means that for these special directions, the matrix acts like a simple multiplier instead of doing complicated rotations.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X, full_matrices=False)  # YOUR CODE HERE

X_reconstructed = U @ np.diag(S) @ Vt  # YOUR CODE HERE — reconstruct from U, S, Vt
X_approx        = S[0] * np.outer(U[:, 0], Vt[0, :])  # YOUR CODE HERE — rank-1 approximation

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**

If we center our data first (by subtracting the average), SVD and PCA are basically doing the same thing. Here's how SVD is translated to PCA:


*   Vt (Right Singular Vectors): The rows of Vt are our Principal Components. This is the matrix that contains the actual directions of maximum variance (the new axes where our data is most spread out).
*   S (Singular Values): These are just numbers that tell us how "important" each of those directions in Vt is. A bigger singular value means the data has a lot of variance in that specific direction.
*   U (Left Singular Vectors): This matrix basically tells us how our original individual data points map onto those new directions.

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)
ks_s1 = stats.kstest(s1, 'norm')  # YOUR CODE HERE
ks_s2 = stats.kstest(s2, 'norm')  # YOUR CODE HERE
print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)  # YOUR CODE HERE — adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts)  # YOUR CODE HERE — np.diff
adf_diff = adfuller(ts_diff)  # YOUR CODE HERE
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

###Interpretation of ADF Test Results (5.4)

* Original Series (p=0.7802): The p-value is much higher than 0.05, so we fail to reject the null hypothesis. This confirms the series is non-stationary, meaning its mean and variance change over time (due to the linear trend).

* Differencing Impact: By subtracting the previous value from each current value (differencing), we successfully removed the trend that was causing the non-stationarity.

* After Differencing (p=0.0000): The p-value dropped to zero, which is below the 0.05 threshold. We now reject the null hypothesis, confirming the series is stationary and ready for time-series modeling.

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    # YOUR CODE HERE
    epsilon=1e-10
    breakpoints = np.linspace(min(expected.min(), actual.min()),
                              max(expected.max(), actual.max()),
                              bins + 1)
    expected_counts, _ = np.histogram(expected, bins=breakpoints)
    actual_counts, _ = np.histogram(actual, bins=breakpoints)
    expected_pct = expected_counts / len(expected) + epsilon
    actual_pct = actual_counts / len(actual) + epsilon
    psi_val = np.sum((actual_pct - expected_pct) * np.log(actual_pct / expected_pct))
    return psi_val
    pass

np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")

# Plot both distributions
fig, ax = plt.subplots()
# YOUR CODE HERE
ax.hist(train_dist, bins=30, alpha=0.5, label='Training Data', color='blue', density=True)
ax.hist(drift_dist, bins=30, alpha=0.5, label='Production (Drifted)', color='red', density=True)
ax.legend(); ax.set_title('Distribution Shift Visualization')
ax.set_xlabel('Feature Value')
ax.set_ylabel('Probability Density')
plt.tight_layout(); plt.show()
````

**Concept Drift vs Covariate Drift & PSI Thresholds (5.5):**

- **Concept drift:** This happens when the actual relationship between  features and the target changes. (Example: People’s buying habits change, so the old model's "logic" no longer works).
- **Covariate drift:** This is when the input data changes, even if the logic stays the same. (Example: The model was trained on 30-year-olds, but now the users are all 60-year-olds).
- **PSI < 0.1:** Stable. The data hasn't really changed; no action is needed.
- **PSI 0.1–0.2:** Minor shift. Should probably investigate or keep a close eye on the model.
- **PSI > 0.2:** Major shift. The data is significantly different; likely need to retrain the model.
- **Retraining trigger example:** If the PSI stays above 0.2 for more than a few days, or if the model's accuracy drops below a certain threshold (like 80%), it’s time to retrain with the new data.

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = (p_free_given_spam * p_spam) + (p_free_given_ham * p_ham)  # YOUR CODE HERE

# Bayes
p_spam_given_free = (p_free_given_spam * p_spam) / p_free  # YOUR CODE HERE

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    # YOUR CODE HERE
    prior_ham = 1 - prior_spam
    evidence = (p_word_given_spam * prior_spam) + (p_word_given_ham * prior_ham)
    posterior = (p_word_given_spam * prior_spam) / evidence
    return posterior
    pass

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

**Bayes Term Mapping (6.3):**

| Term | Formula | In this example |
|------|---------|-----------------|
| Prior | P(Spam) | Our initial belief that an email is spam (30%) before reading it. |
| Likelihood | P(word \| Spam) | If the email is spam, the probability it contains the word "free" (80%). |
| Evidence | P(word) | The total probability of seeing the word "free" in any email (27.5%). |
| Posterior | P(Spam \| word) | Our updated belief that the email is spam (87.3%) now that we've seen the word "free". |

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30
sample_means = np.array([np.random.choice(population, size=sample_size).mean() for _ in range(n_samples)])  # YOUR CODE HERE — array of 5000 sample means

pop_mean = population.mean()
pop_std  = population.std()
clt_std  = pop_std / np.sqrt(sample_size)  # YOUR CODE HERE — predicted std of sample means

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()
# YOUR CODE HERE — histogram of sample means + theoretical normal overlay
ax.hist(sample_means, bins=40, density=True, color='lightgreen', edgecolor='white', alpha=0.7, label='Sample Means')
x_axis = np.linspace(sample_means.min(), sample_means.max(), 100)
y_axis = stats.norm.pdf(x_axis, pop_mean, clt_std)
ax.plot(x_axis, y_axis, color='darkgreen', lw=2, label='CLT Prediction (Normal)')
ax.set_title('CLT: Distribution of Sample Means')
ax.set_xlabel('Mean Value')
ax.set_ylabel('Probability Density')
ax.legend()
plt.tight_layout(); plt.show()

# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

**CLT Reflection (6.4):**

The CLT is important because it proves that the average of any large enough sample will follow a normal distribution, regardless of the original data's shape. This justifies the normality assumption, which is required for classical tests like T-tests and Z-tests to work. It allows researchers to make accurate predictions about a population even when the underlying data is skewed.
